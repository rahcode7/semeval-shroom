```python
import json

data = []
with open(os.path.join(DATA_PATH,DATASET)) as f:
    for line in f:
        data.append(json.loads(line))
df = pd.DataFrame(data)
ic(df.head(4))

df['label_org'] = np.where(df['label']=="Not Hallucination",0,1)
df.head(20)

df['probability_gpt'] = np.where(df['probability_gpt']=="",0,df['probability_gpt'])
```

```python
df.groupby(['task','label_gpt','label_org']).size().reset_index(name='counts')
```
